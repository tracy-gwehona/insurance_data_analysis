```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

```python
# data exploration
insurance_data.info()
```

```python
insurance_data.isna().sum()
```

```python
insurance_data.duplicated().sum()
```

```python
insurance_data_cleaned = insurance_data.copy()
insurance_data_cleaned.drop_duplicates(inplace=True,keep='first')
insurance_data_cleaned
```

```python
insurance_data_cleaned.describe()
```

```python
# correlations
fig, ax = plt.subplots(figsize = (10, 8))

sns.heatmap(insurance_data_cleaned.corr(), annot=True, cmap='coolwarm')

ax.set_title('Correlation matrix')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (20, 16))

sns.barplot(x='age', y='charges', data=insurance_data_cleaned)

ax.set_title('Medical cost charges')
ax.set_xlabel('age')
ax.set_ylabel('charges')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))

sns.boxplot(data=insurance_data_cleaned)

ax.set_title('Box Plot for Insurance Data')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))

sns.scatterplot(data = insurance_data_cleaned)

ax.set_title('Scatter Plot for Insurance Data')

plt.tight_layout()
```
